```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from kmodes.kprototypes import KPrototypes
```

```python
final_data=pd.read_csv('final_csv.csv',low_memory=False)
```

```python
final_data.drop(['Customer No.','Gate Pass Time','Invoice Date', 'Invoice No', 'Invoice Time','Job Card No', 'JobCard Date', 'JobCard Time','Pin code','Recovrbl Exp', 'Regn No','User ID','Unnamed: 0_x','Partner Type','Unnamed: 0_y','DBM Order','Material','Target quantity UoM','Invoice Date Time', 'JobCard Date Time', 'Service Time','Make','Plant Name1','Print Status','Business Partner','District','Model','CITY','Plant','Description'],axis=1,inplace=True)
```

```python
numerical_data=final_data.select_dtypes(include=np.number)
categorical_data=final_data.select_dtypes(exclude=np.number)
```

```python
scalar=StandardScaler()
numerical_data[numerical_data.columns]=scalar.fit_transform(numerical_data[numerical_data.columns])
```

```python
final=pd.concat([numerical_data,categorical_data],axis=1)
```

```python
final.dropna(axis=0,inplace=True)
```

```python
arr=final.values
arr[:,10]=arr[:,10].astype(float)
```

```python
cluster=KPrototypes(n_clusters=4,verbose=2,max_iter=100)
output=cluster.fit_predict(arr,categorical=[11,12,13,14,15,16])
```

```python
cluster.cluster_centroids_
```

```python
cluster_dict=[]
for c in output:
    cluster_dict.append(c)
```

```python
d=np.asarray(cluster_dict)
```

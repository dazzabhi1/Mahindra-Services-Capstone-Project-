```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
customer_rfm = pd.read_csv('../data/customer_rfm.csv')
customer_rfm.head()
```

```python
customer_data = pd.read_csv('../data/customer_data.csv')
```

```python
invoice_data_pincode_cleaned = pd.read_csv('../data/invoice_data_cleaned.csv')
invoice_data_pincode_cleaned.head()
```

```python
invoice_data_pincode_cleaned.shape
```

```python
customer_data.shape
```

```python
#joining customer and invoice data
customer_invoice_right = customer_data.merge(invoice_data_pincode_cleaned, on='Customer No.', how='right', sort=False)
```

```python
customer_invoice_right.shape
```

```python
customer_invoice_right.isnull().sum()
```

```python
jtd = pd.read_csv('../data/JTD.csv')
```

```python
jtd.head()
```

```python
invoice_jtd = jtd.merge(customer_invoice_right, left_on='DBM Order', right_on = 'Job Card No' ,how='right')
```

```python
invoice_jtd.isnull().sum()
```

```python
invoice_jtd.shape
```

```python
invoice_jtd['Invoice DateTime'] = invoice_jtd['Invoice Date']+' '+invoice_jtd['Invoice Time']
invoice_jtd['Jobcard DateTime'] = invoice_jtd['JobCard Date']+' '+invoice_jtd['JobCard Time']

invoice_jtd['Invoice DateTime'] = pd.to_datetime(invoice_jtd['Invoice DateTime'])
invoice_jtd['Jobcard DateTime'] = pd.to_datetime(invoice_jtd['Jobcard DateTime'])

invoice_jtd['Service Time'] = invoice_jtd['Invoice DateTime']-invoice_jtd['Jobcard DateTime']
```

```python
invoice_jtd['Service Time secs'] = invoice_jtd['Service Time'].astype('timedelta64[s]')/86400
```

```python
parts_time = invoice_jtd.groupby(['Description'])['Service Time secs'].mean().reset_index().sort_values(by='Service Time secs', ascending=False)
```

```python
parts_time.drop_duplicates(subset='Description', keep="first", inplace=True)
```

```python
new_cust = pd.read_csv('../data/customer_data.csv')
new_cust.head()
```

```python
#creating dictionaries to replace categorical values in the dataframe with their actual values
partner_type = {'1.0': 'Retail', '2.0': 'Corporate', '3.0': 'Fleet', '4.0': 'Employee', '9001.0': 'Insurance Company', 
                '9002.0': 'Surveyor', '9003.0':'Contact Person'}

data_origin = {
    'Z001': 'Camp-Outdoor',
    'Z002': 'Camp-Workshop',
    'Z003': 'Emailers',
    'Z004': 'Fleet',
    'Z005': 'Reference-Customer',
    'Z006': 'Reference-Employee',
    'Z007': 'Reference-Used Car Dealer',
    'Z008': 'Just Dial/Other helpline',
    'Z009': 'Snapdeal/Other websites',
    'Z010': 'Company website',
    'Z011': 'Float activity',
    'Z012': 'Petrol pump activity',
    'Z013': 'Hoardings/Outdoor ads',
    'Z014': 'Insurance Company',
    'Z015': 'Television Ad',
    'Z016': 'Newspaper Ad',
    'Z017': 'Newspaper leaflet',
    'Z018': 'Outdoor sales activity',
    'Z019': 'Spotted the outlet',
    'Z020': 'Mahindra sister concern employee',
    'Z021': 'Other outdoor activity',
    'Z022': 'Radio'
}

marital_status = {
    '1.0': 'Single',
    '2.0': 'Married',
    '3.0': 'Widowed',
    '4.0': 'Divorced',
    '5.0': 'Separated'
}

occupation = {
    '1.0': 'Director',
    '2.0': 'Sr. Manager',
    '3.0': 'Vice President',
    '4.0': 'Assistant Vice President',
    '5.0': 'Senior Vice President',
    '6.0': 'Col',
    '7.0': 'Lt',
    '8.0': 'Lt. General',
    '9.0': 'Doctor',
    '10.0': 'Captain',
    '11.0': 'Executive Director',
    '12.0': 'Executive Assistant',
    '13.0': 'Secretary',
    '14.0': 'Regional Manager',
    '15.0': 'Manager',
    '16.0': 'Junior Manager',
    '17.0': 'DGM',
    '18.0': 'General Manager',
    '19.0': 'Sr. General Manager',
    '20.0': 'Proprietor',
    '21.0': 'Partner',
    '22.0': 'Vice Chairman',
    '23.0': 'CFO',
    '24.0': 'CTO',
    '25.0': 'Self',
    '26.0': 'Managing Director',
    '27.0': 'Executive',
    '28.0': 'CEO',
    '29.0': 'Customer',
    '30.0': 'VIP Customer',
    '31.0': 'Others'
}

title = {
    '1.0': 'Ms.',
    '2.0': 'Mr.',
    '3.0': 'Company',
    '4.0': 'Mr. and Ms.'
}
```

```python
new_cust['Partner Type'] = new_cust['Partner Type'].astype('str')
new_cust['Title'] = new_cust['Title'].astype('str')
new_cust['Marital Status'] = new_cust['Marital Status'].astype('str')
new_cust['Occupation'] = new_cust['Occupation'].astype('str')
```

```python
new_cust.replace({'Partner Type': partner_type}, inplace=True)
new_cust.replace({'Data Origin': data_origin}, inplace=True)
new_cust.replace({'Title': title}, inplace=True)
new_cust.replace({'Marital Status': marital_status}, inplace=True)
new_cust.replace({'Occupation': occupation}, inplace=True)

new_cust['Partner Type'].replace({'nan': 'Unknown'}, inplace=True)
new_cust['Data Origin'].replace({'nan': 'Unknown'}, inplace=True)
new_cust['Title'].replace({'nan': 'Unknown'}, inplace=True)
new_cust['Marital Status'].replace({'nan': 'Unknown'}, inplace=True)
new_cust['Occupation'].replace({'nan': 'Unknown'}, inplace=True)


new_cust.head()
```
